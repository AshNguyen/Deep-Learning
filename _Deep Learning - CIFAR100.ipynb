```python
from keras.datasets import cifar100
import keras as kr
from keras.layers import Dense, Activation



(X_train, y_train), (X_test, y_test) = cifar100.load_data(label_mode='fine')

X_train, X_test = X_train/255, X_test/255
y_train = kr.utils.to_categorical(y_train, 100)
y_test = kr.utils.to_categorical(y_test, 100)
X_train_rs, X_test_rs = X_train.reshape([50000, 32*32*3]), X_test.reshape([10000, 32*32*3])
```

```python
import matplotlib.pyplot as plt

def cifar_pic(image): 
    plt.imshow(image, interpolation="nearest")
    plt.axis('off')
    plt.show()
    

def plot_cifar(x,y ,images): 
    for _ in range(images.shape[0]):
        plt.subplot(x, y, _+1)
        plt.imshow(images[_], interpolation="nearest")
        plt.axis("off")
    plt.show()
    
plot_cifar(5,5,X_train[0:25])
```

# Dense Net

```python
model=kr.Sequential()
model.add(Dense(units=512, input_shape=(32*32*3,), activation='relu'))
model.add(Dense(units=300, activation='relu'))
model.add(Dense(units=100, activation='softmax'))
```

```python
import keras.backend as K
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=20, batch_size=400, verbose=1)
```

```python
model.save('/Users/ash/Downloads/cifar100.ash')
```

```python
scores=model.evaluate(X_test, y_test)
print('Accuracy:', scores[1])
```
