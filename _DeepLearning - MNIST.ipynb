# Fetch And Prepare MNIST Data 

```python
from sklearn.datasets import fetch_mldata 
data=fetch_mldata('MNIST original')
X, y = data["data"], data["target"]
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
import numpy as np
shuffle_index = np.random.permutation(60000)
X_train, y_train = X_train[shuffle_index], y_train[shuffle_index] 
y_train = kr.utils.to_categorical(y_train, 10)
y_test = kr.utils.to_categorical(y_test, 10)
X_train = X_train/255
X_test = X_test/255
```

# Define Dense Neural Net With Different Activation

```python
import keras as kr
from keras.layers import Dense, Dropout
from keras.callbacks import TensorBoard

board= TensorBoard(log_dir='/Users/ash/Downloads/logs', histogram_freq=0,
                          write_graph=True, write_images=False)
```

```python
neuralNet= kr.Sequential()
neuralNet.add(Dense(512, input_shape=(784,), activation='relu'))
neuralNet.add(Dropout(0.6))
neuralNet.add(Dense(512, activation='relu'))
neuralNet.add(Dropout(0.6))
neuralNet.add(Dense(10, activation='softmax'))
```

```python
neuralNet.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
neuralNet.fit(X_train, y_train, epochs=20, batch_size=200, verbose=1, callbacks=[board])
```

```python
scores=neuralNet.evaluate(X_test, y_test, verbose=0)
print('Accuracy', scores[1])
print('Loss', scores[0])
```

```python
neuralNet_2= kr.Sequential()
neuralNet_2.add(Dense(512, input_shape=(784,), activation='sigmoid'))
neuralNet_2.add(Dense(512, activation='relu'))
neuralNet_2.add(Dense(10, activation='softmax'))

neuralNet_2.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
neuralNet_2.fit(X_train, y_train, epochs=20, batch_size=200, verbose=0)

scores_2=neuralNet_2.evaluate(X_test, y_test, verbose=0)
print('Accuracy', scores_2[1])
print('Loss', scores_2[0])
```

```python
neuralNet_3= kr.Sequential()
neuralNet_3.add(Dense(512, input_shape=(784,), activation='sigmoid'))
neuralNet_3.add(Dense(512, activation='sigmoid'))
neuralNet_3.add(Dense(10, activation='softmax'))

neuralNet_3.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
neuralNet_3.fit(X_train, y_train, epochs=20, batch_size=200, verbose=0)

scores_3=neuralNet_3.evaluate(X_test, y_test, verbose=0)
print('Accuracy', scores_3[1])
print('Loss', scores_3[0])
```
